```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time
```

```python
# Set up Selenium with Chrome
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
# Open the URL
url = "https://www.uptodate.com/contents/search?search=colonoscopy&sp=0&searchType=PLAIN_TEXT&source=USER_INPUT&searchControl=TOP_PULLDOWN&autoComplete=false"
driver.get(url)
```

```python
while True:
    try:
        # Locate the "Show More Results" button
        show_more_button = driver.find_element(By.ID, "showmore-button")
        
        # Click the button to load more results
        show_more_button.click()
        
        # Wait briefly for new results to load
        time.sleep(20)
        
    except NoSuchElementException:
        # If the button is not found, it means all results have loaded
        print("No more 'Show More Results' button available. All results loaded.")
        break
```

```python
# After loading all results, retrieve the page content
expanded_content = driver.page_source
# print beautifulsoup
soup = BeautifulSoup(expanded_content, 'html.parser')
print(soup.prettify())
```

```python
# Locate all list items with role="listitem" and extract links
list_items = driver.find_elements(By.CSS_SELECTOR, 'li[role="listitem"] a.search-result-primary-link')
links = [item.get_attribute("href") for item in list_items]

# Print all extracted links
for link in links:
    print(link)
```

```python
# the links contain search query, which is sessional. 
# We need to remove such query to get the correct link
import re
import urllib
from urllib.parse import urlparse, urlunparse, parse_qs, urlencode
pattern_to_split = r"search=colonoscopy&source"

def split_url(url, pattern):
    n_url = url.split(pattern)
    return n_url[0]

# split the url
n_links = [split_url(link, pattern_to_split) for link in links]
n_links
```

```python
# save the links to a file
with open("UpToDate_colonoscopy_links.txt", "w") as file:
    for link in links:
        file.write(link + "\n")
```

```python
# List to store the scraped article data
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time
import random

# Initialize variables
article_content = []
session_limit = 20  # Number of requests per session
count = 0
driver = webdriver.Chrome()  # Start browser session

for link in n_links:
    count += 1
    driver.get(link)
    
    # Retry mechanism
    retries = 3  # Number of retries
    for attempt in range(retries):
        try:
            # Wait until the article tag is present (up to 10 seconds)
            article = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.TAG_NAME, "article"))
            )
            article_text = article.text
            
            # Store the article data in a dictionary
            article_content.append({
                "url": link,
                "article_text": article_text
            })
            
            print(f"{count} - Scraped article from {link}")
            break  # Exit the retry loop if successful

        except (TimeoutException, NoSuchElementException):
            print(f"Article content not found for {link}, attempt {attempt + 1} of {retries}")
            if attempt < retries - 1:
                driver.refresh()  # Refresh the page and try again
            else:
                print(f"Failed to scrape article after {retries} attempts for {link}")
    
    # Restart session after reaching session limit
    if count % session_limit == 0:
        driver.quit()  # Close the browser session
        wait_time = random.uniform(30, 60)  # Wait between 30 and 60 seconds before new session
        print(f"Restarting session. Waiting {wait_time:.2f} seconds.")
        time.sleep(wait_time)
        driver = webdriver.Chrome()  # Start a new browser session

# Close the final session
driver.quit()


```

```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import ElementClickInterceptedException, NoSuchElementException
import time

# Initialize the WebDriver once at the start of your script
driver = webdriver.Chrome()  # Make sure `driver` is initialized only once

def auto_show_more(driver, url):
    # Navigate to the specified URL in the current window
    driver.get(url)
    
    while True:
        try:
            # Locate the "Show More Results" button
            show_more_button = driver.find_element(By.ID, "showmore-button")
            
            # Scroll to the button if it's out of view
            driver.execute_script("arguments[0].scrollIntoView();", show_more_button)
            
            # Wait briefly for the button to be in view
            time.sleep(1)
            
            try:
                # Click the button to load more results
                show_more_button.click()
                # Wait briefly for new results to load
                time.sleep(3)
            
            except ElementClickInterceptedException:
                # Handle case where another element is in front of the button
                print("Element obstructed, attempting to click again.")
                ActionChains(driver).move_to_element(show_more_button).click().perform()

        except NoSuchElementException:
            # If the button is not found, all results have likely loaded
            print("No more 'Show More Results' button available. All results loaded.")
            break



# Don't forget to close the driver when you're done
# driver.quit()
```

```python
# Now we query all links that presented
from selenium.webdriver.common.by import By
from selenium import webdriver

def extract_links():
    """
    Extracts links from list items with role="listitem" on the current page.
    
    Args:
        driver (webdriver): The WebDriver instance currently in use.
    
    Returns:
        list: A list of URLs extracted from the page.
    """
    # Locate all list items with role="listitem" and extract links
    list_items = driver.find_elements(By.CSS_SELECTOR, 'li[role="listitem"] a.search-result-primary-link')
    links = [item.get_attribute("href") for item in list_items]
    
    # Print all extracted links (optional)
    for link in links:
        print(link)
    
    return links

# Usage example
# driver = webdriver.Chrome()  # Ensure driver is initialized only once in your script
# driver.get("https://example.com")  # Replace with the URL of the page to extract links from
gastro_links = extract_links()
```

```python
# split the links by pattern ?search=Gastrointestinal+Disorder, and keep the first part
pattern_to_split = r"search=Gastrointestinal+Disorders&source"

def split_url(url, pattern):
    n_url = url.split(pattern)
    return n_url[0]

# split the url
n_gastro_links = [split_url(link, pattern_to_split) for link in gastro_links]
n_gastro_links
```
